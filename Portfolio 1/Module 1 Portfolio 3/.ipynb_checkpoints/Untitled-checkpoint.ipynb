```python
import csv
dataset = list(csv.reader(open("stu.csv",encoding = "utf-8")))
```

```python
dataset
```

```python
dataset[0].append("age_column")
```

```python
dataset[0].index("age")
```

```python
for i in dataset[1:]:
    age = int(i[3])
    if age <= 12:
        i.append("Child")
    elif age <= 20:
        i.append("Teenager")
    elif age <= 50:
        i.append("Adult")
    elif age <=  100:
        i.append("Senior Citizen")
    else:
        i.append("Error")
```

```python
## to dinf how many age group
freq = {}
for i in dataset[1:]:
    age_group = i[-1]
    if age_group in freq:
        freq[age_group] += 1
    else:
        freq[age_group] = 1
freq
```

```python
## to find how many race
dataset[0].index("race")
```

```python
freq_race = {}
for i in dataset[1:]:
    race = i[4]
    if race in freq_race:
        freq_race[race] += 1
    else:
        freq_race[race] = 1
freq_race
```

```python
freq_gender = {}
for i in dataset[1:]:
    gender = i[2]
    if gender in freq_gender:
        freq_gender[gender] += 1
    else:
        freq_gender[gender] = 1
freq_gender
```
