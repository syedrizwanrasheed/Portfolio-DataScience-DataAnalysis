```python
import csv
dataset = list(csv.reader(open("AppleStore.csv",encoding = "utf-8")))
```

```python
dataset[0]
```

```python
### Frequency table for genre 
for i in dataset[1:]:
    genre = i[11]
    print (genre)
```

```python
freqprimegenre = {}
for i in dataset[1:]:
    genre = i[11]
    if genre not in freqprimegenre:
        freqprimegenre[genre] = 1
    else:
        freqprimegenre[genre] += 1
freqprimegenre
```

```python
## frequency of track name

dataset[0].index("track_name")
```

```python
tracklist = {}
for i in dataset[1:]:
    trackname = i[1]
    if trackname not in tracklist:
        tracklist[trackname] = 1
    else:
        tracklist[trackname] += 1 
tracklist
```

```python
### frequency of price

dataset[0].index("price")
```

```python
price_freq = {}
for i in dataset[1:]:
    price_list = i[4]
    if price_list not in price_freq:
        price_freq[price_list] = 1
    else:
        price_freq[price_list] += 1
price_freq
```

```python
dataset[0].index("currency")
```

```python
currency_freq = {}
for i in dataset[1:]:
    currency_list = i[3]
    if currency_list not in currency_freq:
        currency_freq[currency_list] = 1
    else:
        currency_freq[currency_list] += 1
currency_freq
            
```

```python
### adding column of user labelling
dataset[0].append("price labelling")
```

```python
for i in dataset[1:]:
    price_list = float(i[4])
    if price_list == 0:
        i.append("free")
    elif price_list > 0 and price_list <= 20:
        i.append("cheap")
    elif price_list > 20 and price_list <= 50:
        i.append("affordable")
    elif price_list > 50:
        i.append("expensive")
    else:
        i.append("price not found")
        
```

```python
### adding frequency tabel of user_labelling clomun and data in dataset
dataset[0].append("user_rate_labelling")
```

```python
for i in dataset[1:]:
    user_rating = float(i[7])
    if user_rating == 0:
        i.append("poor")
    elif user_rating > 0 and user_rating <=2:
        i.append("average")
    elif user_rating > 2 and user_rating <= 4:
        i.append("Good")
    elif user_rating > 4:
        i.append("Awesome")
    else:
        i.append("Rating not found")
```
