```python
covid_country_quantity = {}
for i in covid[1:]:
    covid_country_reported_quantity = i[1]
    if covid_country_reported_quantity in covid_country_quantity:
        covid_country_quantity[covid_country_reported_quantity] += 1
    else:
        covid_country_quantity[covid_country_reported_quantity] = 1
```
