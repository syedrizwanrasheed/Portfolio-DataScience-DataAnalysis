```python
freq_user = []
for i in body:
    username = i[3]
    if username not in freq_user:
        freq_user.append(username)
```

```python
freq_Id = []
for i in body:
    Id = i[1]
    if Id not in freq_Id:
        freq_Id.append(Id)
```

```python
freq_text = []
for i in body:
    text = i[2]
    if text not in freq_text:
        freq_text.append(text)
```

```python
from datetime import datetime
```

```python
for i in body:
    i[0] = i[0].strip("TIME")
```

```python
format_data = "%Y-%m-%d %H:%M:%S"
for i in body:
    i[0] = datetime.strptime(i[0] , format_data)
    print(type(i[0]))
```

```python
for i in body:
    print (i[0].month)
```

```python
day_freq = {}
for i in body:
    time = i[0]
    hours = time.strftime("%I %p")
    if hours not in day_freq:
        day_freq[hours] = 1
    else:
        day_freq[hours] += 1
day_freq
```
