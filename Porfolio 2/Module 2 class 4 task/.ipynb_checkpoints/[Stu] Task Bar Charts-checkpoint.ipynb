```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
medal = pd.read_csv("medals_by_country_2016.csv")
```

```python
medal
```

```python
medal.rename(columns={'Unnamed: 0':"Country"}, inplace = True)
medal
```

```python
#for bronze
x = medal["Country"]
y = medal["Bronze"]
```

```python
#plt.style.use("fivethirtyeight")
plt.bar(x,y,linestyle = '-' , color = 'b')
plt.title('Bronze Medal Frequency')
plt.xlabel('Country')
plt.ylabel('Medal Counts')
plt.bar(x,y)
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
x_2 = medal["Country"]
y_2 = medal["Gold"]
```

```python
plt.bar(x_2 , y_2)
plt.title("Gold medal frequency")
plt.xlabel("Country")
plt.ylabel("Medal Counts")
plt.bar(x_2,y_2,linestyle = "-" , color = "g")
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
# for silver
x_3 = medal["Country"]
y_3 = medal["Silver"]
plt.bar(x_3,y_3)
plt.title("Silver medal counts")
plt.xlabel("Country")
plt.ylabel("Silver")
plt.bar(x_3,y_3,linestyle = '-' , color = '#8a3246')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
medal.plot.bar(x='Country', stacked=True, title='Medal Counts', color = ['olive','cyan','pink'], ylabel=('Medal count'))
```
