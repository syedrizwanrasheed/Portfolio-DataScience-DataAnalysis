```python
import pandas as pd
```

```python
laptops = pd.read_csv("laptops.csv" , encoding = 'latin-1')
laptops.head()
```

```python
laptops.columns
```

```python
laptops.columns = laptops.columns.str.replace("(" , "").str.replace(")" , "").str.strip(" ")
```

```python
laptops.columns
```

```python
laptops.rename({'Operating System' : 'OS'} , axis = 1 , inplace = True)
```

```python
laptops.columns
```

```python
laptops.columns = laptops.columns.str.replace(" " , "_")
```

```python
laptops.rename({'RAM' : 'RAM_GB' \
                , 'Screen_Size' : 'Screen_Size_Inches'} , axis = 1 , inplace = True)
laptops.columns
```

```python
laptops['RAM_GB'] = laptops['RAM_GB'].str.strip("GB").astype(float)
laptops['RAM_GB']
```

```python
laptops["Screen_Size_Inches"] = laptops["Screen_Size_Inches"].str.replace('"' , "").astype(float)
```

```python
laptops.columns
```

```python
laptops.loc[: , "CPU"]
```

```python
laptops.loc[: , 'CPU'].str.split(" " , n = 1 , expand = True) [0]
```

```python
Cpu_Brand = laptops.loc[: , 'CPU'].str.split(" " , n = 1 , expand = True) [0]
```

```python
laptops.insert(6 , "Cpu_Brands" , Cpu_Brand)
```

```python
laptops.head()
```

```python
laptops.head()
```

```python
laptops.loc[: , "GPU"]
```

```python
Gpu_Brand = laptops.loc[: , "GPU"].str.split(" " , n = 1 , expand = True) [0]
```
