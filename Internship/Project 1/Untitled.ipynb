```python
max_reviews = App_net.groupby('App')['Reviews'].max()
```

```python
app_filtered = []
app_raw = []
for index, row in App_net.iterrows():
    name = row["App"]
    name_of_reviews = row["Reviews"]
    if (max_reviews[name] == name_of_reviews) and (name not in app_raw):
        app_filtered.append(row)
        app_raw.append(name)
```

```python
android_final["Installs_int"] = android_final["Installs"].str.replace("," , "").str.replace("+","").str.replace("+","").astype(int)
```

```python
# we want to find install freater than 1000 per app
```

```python
install_frq = android_final["Installs_int"].value_counts().sort_index()
install_frq = install_frq[install_frq.index > 500]
install_frq
```

```python
def alphanymeric_units(value):
    if value>= 1e9:
        return f'{value/1e9:.0f}B' 
    elif value>= 1e6:
        return  f'{value/1e6:.0f}M'
    elif value>= 1e3:
        return f'{value/1e3:.0f}K'
    else:
        return f'{value:.0f}'
```

```python
install_frq.index = install_frq.index.map(alphanymeric_units)
```

```python
install_frq.index
```
