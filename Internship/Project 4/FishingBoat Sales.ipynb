```python
# setting plot style
plt.style.use('dark_background')
# setting axis
x = FishingBoat["Manufacturer"]
y = FishingBoat["Number of views last 7 days"]
# setting figsize
fig, ax = plt .subplots(figsize = (10,20)) #width , length
# Horizontal Bar plot
ax.barh(x , y , color = "#2ED4A5")
# Adding grid
ax.grid(color = 'grey' , alpha = 0.5)
# Removing spines/border
pos = ["top" , "bottom" , "left" , "right"]
for i in pos:
    #print(i)
    ax.spines[i].set_visible(False)
# Removing ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Sorting In ascending order
ax.invert_yaxis()

# Adding title
ax.set_title("Boat Company Sales View" , loc = 'center',\
            fontdict = {'family' : 'monospace' , 'size' : 50 , 'weight':'bold'})
# Displaying Graph
plt.show
```
